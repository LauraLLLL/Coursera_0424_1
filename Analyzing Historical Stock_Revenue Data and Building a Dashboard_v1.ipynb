Question 2: Use Webscraping to Extract Tesla Revenue Data

Display the last five rows of the tesla_revenue dataframe using the tail function. Upload a screenshot of the results.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    tesla_revenue = pd.read_html(str(soup), match="Tesla Quarterly Revenue")[0]

    # Rename columns for better readability
    tesla_revenue.columns = ['Date', 'Quarterly Revenue (Millions USD)']

    # Replace NaN values with a placeholder to indicate missing data
    tesla_revenue['Quarterly Revenue (Millions USD)'] = tesla_revenue['Quarterly Revenue (Millions USD)'].fillna('Data not available')

    # Display the last five rows of the dataframe with a more formatted output
    print(tesla_revenue.tail().to_string(index=False))
else:
    print("Failed to retrieve content: Status code", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html5lib")
gme_revenue_tables = soup.find_all("table")
gme_revenue_table = None
for index, table in enumerate(gme_revenue_tables):
    if "GameStop Quarterly Revenue" in str(table):
        gme_revenue_table = table

#gme_revenue = pd.read_html(str(gme_revenue_table), flavor= "bs4")[0]
#gme_revenue.columns = ["Date", "Revenue"]

gme_revenue = pd.DataFrame(columns = ["Date", "Revenue"])

for row in gme_revenue_table.tbody.find_all("tr"):
    col = row.find_all("td")
    if col != []:
        date = str(col[0].string)
        revenue = str(col[1].string)
        gme_revenue = pd.concat([gme_revenue, pd.DataFrame({"Date": [date], "Revenue": [revenue]})], ignore_index=True)


gme_revenue["Revenue"] = gme_revenue['Revenue'].str.replace(',|\$',"", regex=True)

gme_revenue.dropna(inplace=True)
gme_revenue = gme_revenue[gme_revenue['Revenue'] != "None"]
gme_revenue.tail()
```

Question 5: Plot Tesla Stock Graph

Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.

Upload a screenshot of your results.

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import requests
from bs4 import BeautifulSoup

def make_graph(stock_data, revenue_data, title):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))  # Create two subplots

    # First subplot for the closing stock price
    ax1.plot(stock_data['Date'], stock_data['Close'], label='Closing Price', color='blue')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Closing Price (USD)', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    ax1.set_title(title + ' - Historical Share Price')
    ax1.legend(loc='upper left')

    # Second subplot for the revenue
    ax2.plot(revenue_data['Date'], revenue_data['Quarterly Revenue (Millions USD)'], label='Quarterly Revenue', color='red')
    ax2.set_xlabel('Date')
    ax2.set_ylabel('Revenue (Millions USD)', color='red')
    ax2.tick_params(axis='y', labelcolor='red')
    ax2.set_title(title + ' - Historical Revenue')
    ax2.legend(loc='upper left')

    # Automatically adjust subplot params for better layout
    plt.tight_layout()

    # Show the plot
    plt.show()

try:
    # Extracting stock data for Tesla
    tesla_stock_data = yf.download('TSLA', start='2010-01-01', end='2024-04-23')
    tesla_stock_data.reset_index(inplace=True)

    # Extracting revenue data
    url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'
    }

    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        tesla_revenue_table = pd.read_html(str(soup), match="Tesla Quarterly Revenue")[0]
        tesla_revenue_table.columns = ['Date', 'Quarterly Revenue (Millions USD)']
        tesla_revenue_table['Date'] = pd.to_datetime(tesla_revenue_table['Date'])
        tesla_revenue_table['Quarterly Revenue (Millions USD)'] = pd.to_numeric(
            tesla_revenue_table['Quarterly Revenue (Millions USD)']
            .str.replace(',', '')
            .str.replace('$', ''), errors='coerce'
        )
        
        # Call the function to make the graph
        make_graph(tesla_stock_data, tesla_revenue_table, 'Tesla')

except Exception as e:
    print("An error occurred:", e)
```
